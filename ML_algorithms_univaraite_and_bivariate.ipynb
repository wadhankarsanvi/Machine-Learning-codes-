 the percentage of passengers who survived in each of the three passenger classes.

 titanic.groupby('Pclass'): This groups the DataFrame by the unique values in the 'Pclass' column (1, 2, and 3).
['Survived']: After grouping, this selects the 'Survived' column within each group.
.mean(): This calculates the mean of the 'Survived' column for each group. Since 'Survived' is a binary variable (0 or 1), the mean represents the proportion of survivors in each passenger class.
* 100: This multiplies the mean by 100 to express the survival rate as a percentage.

```python
(titanic.groupby('Pclass')['Survived'].mean())*100
#we want to group by the mention classs in this case we are getting if the mention classed is survived or not by getting their average
```

A clustered heatmap also performs hierarchical clustering on the rows and columns to group similar categories together, which can help reveal patterns in the data.
